```python
# Task 2.1

class Node:
    def __init__(self, data):  # constructor
        self.data = data
        self.left = None
        self.right = None

    def get_data(self):   # getter for data
        return self.data

    def get_left(self):  # getter for left
        return self.left

    def get_right(self):  # getter for right
        return self.right

    def set_left(self, left):  # setter for left
        self.left = left

    def set_right(self, right):  # setter for right
        self.right = right

    
```

```python
# Task 2.2

path = input('File path: ')

file = open(path, 'r')

chardict = {}

for line in file:
    for char in line:
        if char in chardict.keys():  # check if the character is alreaady in the dictionary
            chardict[char] += 1  # if yes, add 1 to the count
        else:
            chardict[char] = 0  # else initialise the key-value pair

file.close()

# temporarily using nodelist to store chardict as a list of tuples for easy sorting
nodelist = [(chardict[char], char) for char in chardict]
# sort nodelist
nodelist.sort()
# convert all elements to type Node
for i in range(len(nodelist)):
    nodelist[i] = Node([nodelist[i][1], nodelist[i][0]])  # add to nodelist. the last element is the frequency
```

```python
# Task 2.3

while len(nodelist) > 1:
    # find lowest and 2nd lowest
    freqlist = [nodelist[i].get_data()[-1] for i in range(len(nodelist))]  # list of frequencies
    lowest = freqlist.index(min(freqlist))  # Index of lowest node
    freqlist[lowest] = max(freqlist)  # remove lowest node's frequency
    second = freqlist.index(min(freqlist))  # Index of second lowest node

    # create a temporary variable for the new node
    temp = Node(
        nodelist[lowest].get_data()[:-1] + nodelist[second].get_data()[:-1] +
        [nodelist[lowest].get_data()[-1] + nodelist[second].get_data()[-1]])
    temp.set_left(nodelist[lowest])
    temp.set_right(nodelist[second])

    # use nodelist to store temp
    nodelist[lowest] = temp
    nodelist.pop(second)

# tree is the top node and also the tree
tree = nodelist[0]
```

```python
# Task 2.4

# recursive function to find every single character
def getbit(node):  # returns dict
    bits = {}
    if node.left != None:
        # left
        leftbits = getbit(node.get_left())  # get previous instances
        for k in leftbits:
            leftbits[k] = '0'+leftbits[k]  # add 0 to string
        bits.update(leftbits)  # merge leftbits
    if node.right != None:
        # right
        rightbits = getbit(node.get_right())  # get previous instances
        for k in rightbits:
            rightbits[k] = '1'+rightbits[k]  # add 1 to string
        bits.update(rightbits)  # merge rightbits

    if node.get_left() == None and node.get_right() == None:
        return {node.get_data()[0]: ''}  # no path
    else:
        return bits  # return the final list

bits = getbit(tree)

# open files
readf = open(path, 'r')
writef = open('HAMLET_compressed.txt', 'w')

finalwrite = '' # final string to write to the file
for line in readf:
    for char in line:
        finalwrite = finalwrite + bits[char]  # append to the end of the final string
writef.write(finalwrite)

# close files
readf.close()
writef.close()
```
